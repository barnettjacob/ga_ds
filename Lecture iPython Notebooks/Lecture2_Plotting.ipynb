```python
x=linspace(-5,5,20)
```

```python
#Hello this is a linear space from -5 to 5 with 10 points
```

```python
x
```

```python
y=x**3
y
```

```python
figure()
plot(x, y,'r')
xlabel('x')
ylabel('y')
```

```python
subplot(1,2,1)
plot(x,y,'b--')
subplot(1,2,2)
plot(y,x,'k*-')
```

```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, 0.8, 0.8])
axes.plot(x,y,'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
```

```python
(fig, axes) = plt.subplots(nrows=1, ncols=2)

for ax in axes: 
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('My plot')

```

```python
# A histogram
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("Cumulative detailed histogram")
axes[1].set_xlim((min(n), max(n)));
```
